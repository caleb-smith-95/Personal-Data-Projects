```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import urllib
```

```python
def dk_scrape(sport, league, category, subcategory):
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    
    url = f'https://sportsbook.draftkings.com/leagues/{sport}/{league}?intendedSiteExp=US-IL-SB&category={category}&subcategory={subcategory}'
    headers={'User-Agent':user_agent,}
    
    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
    
    return pd.read_html(data)
```

```python
def fd_scrape(league, tab):
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    
    url = f'https://sportsbook.fanduel.com/navigation/{league}?tab={tab}'
    headers={'User-Agent':user_agent,}
    
    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
    
    return pd.read_html(data)
```

```python
gamedays = []

for i in range(len(game_lines)):
    table = game_lines[i]
    df = pd.DataFrame(table)
    gamedays.append(df)
```

```python
for g in range(len(gamedays)):
    gamedays[g]['Date'] = gamedays[g].columns[0]
    gamedays[g].columns.values[0] = 'Team' # Renames first column only
```

```python
gamedays[0]
```

```python
all_lines = pd.DataFrame()

for g in range(len(gamedays)):
    all_lines = pd.concat([all_lines, gamedays[g]], ignore_index = True, axis = 0)
```

```python
all_lines.reset_index(inplace = True)
all_lines
```

```python
away = all_lines[all_lines.index % 2 == 0]
away.rename(columns = {'index': 'Away Index', 'Team': 'Away Team', 'Spread': 'Away Spread', 'Total': 'Over', 'Moneyline': 'Away Moneyline'}, inplace = True)
away.reset_index(inplace = True, drop = True)
away
```

```python
home = all_lines[all_lines.index % 2 == 1]
home.rename(columns = {'index': 'Home Index', 'Team': 'Home Team', 'Spread': 'Home Spread', 'Total': 'Under', 'Moneyline': 'Home Moneyline'}, inplace = True)
home.drop(columns = ['Date'], inplace = True)
home.reset_index(inplace = True, drop = True)
home
```

```python
full_df = pd.concat([home, away], axis = 1)
full_df.dropna(inplace = True)
full_df.reset_index(inplace = True, drop = True)
full_df
```

```python
full_df['Home Team'].str.split()[0][1]
```

```python
home_teams = full_df['Home Team'].str.split().apply(lambda x: x[1])
home_teams
```

```python
away_teams = full_df['Away Team'].str.split().apply(lambda x: x[1])
away_teams
```

```python
under = full_df['Under'].str.split().apply(lambda x: x[1])
under_line = under.str.split('−').apply(lambda x: x[0])
under_odds = under.str.split('−').apply(lambda x: x[1])
under_odds
```

```python
over = full_df['Over'].str.split().apply(lambda x: x[1])
over_line = over.str.split('−').apply(lambda x: x[0])
over_odds = over.str.split('−').apply(lambda x: x[1])
over_line
```

```python
away_spread = full_df['Away Spread'].str.split('−|\+').apply(lambda x: x[-2:])
away_points = away_spread.apply(lambda x: x[0])
away_odds = away_spread.apply(lambda x: x[1])
away_odds
```

```python
away_points
```

```python
clean_df = pd.DataFrame({'Home Team': home_teams, 'Away Team': away_teams, 'Home Moneyline': full_df['Home Moneyline'], 'Away Moneyline': full_df['Away Moneyline'], 'Home Spread': home_points, 'Home Spread Odds': home_odds, 'Away Spread': away_points, 'Away Spread Odds': away_odds, 'Over/Under': over_line, 'Over Odds': over_odds, 'Under Odds': under_odds})
clean_df.replace('\+', '', regex=True, inplace = True)
clean_df.replace('\−', '-', regex = True, inplace = True)
clean_df['Home Moneyline'] = clean_df['Home Moneyline'].astype(float)
clean_df['Away Moneyline'] = clean_df['Away Moneyline'].astype(float)
clean_df['Home Spread'] = clean_df['Home Spread'].astype(float)
clean_df['Away Spread'] = clean_df['Away Spread'].astype(float)
clean_df['Home Spread Odds'] = clean_df['Home Spread Odds'].astype(float)
clean_df['Away Spread Odds'] = clean_df['Away Spread Odds'].astype(float)
clean_df['Over/Under'] = clean_df['Over/Under'].astype(float)
clean_df['Over Odds'] = clean_df['Over Odds'].astype(float)
clean_df['Under Odds'] = clean_df['Under Odds'].astype(float)
clean_df
```
