```python
brady = pd.DataFrame()
```

```python
url = "https://www.pro-football-reference.com/players/B/BradTo00.htm"
data = pd.read_html(url)
brady = data[1]
brady.drop(index = [23, 24, 25], inplace = True)
brady
```

```python
brady['Age'] = pd.to_numeric(brady.Age)
brady_new = brady[brady.Age >= 30]
brady_new
```

```python
cmp = brady_new['Cmp'].sum()
att = brady_new['Att'].sum()
INT = brady_new['Int'].sum()
td = brady_new['TD'].sum()
yds = brady_new['Yds'].sum()

a = 5*(cmp/att - 0.3)
b = 0.25*(yds/att - 3)
c = 20 * td/att
d = 2.375 - 25*(INT/att)

rating = 100*(a+b+c+d)/6
rating
```
